```python
import pandas as pd
from pandas.api.types import CategoricalDtype
```

```python
df = pd.read_excel("Why Do Students Withdraw From Courses_ (Responses) austin (1).xlsx")
df
```

```python
len("Please rate how much you agree or disagree with the following statements: I dropped the course because I [")
```

```python
cols = list(df.columns)
newcols = []
for name in cols:
    if name[:6] == "Please":
        name = name[106:-1]
    newcols.append(name)
newcols
```

```python
df.columns = newcols
df
```

```python
df.dtypes
```

```python
agreeCols = list(df.columns)[7:-1]
agreeCols
```

```python
cat_type = CategoricalDtype(categories=["Strongly Disagree", "Disagree", "Neutral", "Agree", "Strongly Agree"], ordered=True)
for col in agreeCols:
    df[col] = df[col].astype(cat_type)
```

```python
convert = {}
for col in agreeCols:
    convert[col] = "category"
df = df.astype(convert)
df.dtypes
```

```python
for col in agreeCols:
    df[col] = df[col].cat.codes
df
```

```python
for col in agreeCols:
    print((df[col] > 3).sum(), col)
```

```python
for col in agreeCols:
    print((df[col] % 2 == 1).sum(), col)
```

```python
for col in agreeCols:
    print((df[col] % 2 == 0).sum(), col)
```
